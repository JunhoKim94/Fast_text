```python
import numpy as np
import pandas as pd
```

```python
path = "./data/ag_news/train.csv"
```

```python
data = pd.read_csv(path, header = None, names = ["class", "title", "description"])
```

```python
x = np.array(data)
```

```python
words = []
for line in x:
    words += line[1].split()
    words += line[2].split()
```

```python
word2idx = {"UNK" : 0}
for word in words:
    if word not in word2idx:
        word2idx[word] = len(word2idx)
```
